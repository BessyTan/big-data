```python
# Import and initialize findspark for pyspark installation
# import pyspark module, setting up a Spark application
# import SparkContext to interface with Spark clusters, SparkConf to configure Spark
import findspark
findspark.init()
import pyspark
from pyspark import SparkContext, SparkConf
```

```python
# Create a SparkConf object to configure the Spark application
    # name "Extraction Transformation" 
    # master URL "local[*]" so Spark will run locally
# create entry points to spark
conf = SparkConf().setAppName("Extraction Transformation").setMaster("local[*]")
sc = SparkContext(conf=conf)
```

```python
# delete the directory '/notebooks/wordCountNOTTICE_Homework3'
# read file "/spark/NOTICE" and return an RDD, number of partitions is '1')
# flatmap return one word per element, not a list
# split each line into words by spaces
# map create a tuple for each word in the form (word, 1) to count the occurrences of each word
# reduceByKey reduce the data by summing up the total word count for each unique word
# saveAsTextFile save the word count result to file "/notebooks/wordCountNOTICE_homework3"
! rm -rf /notebooks/wordCountNOTICE_homework3
sc.textFile("/spark/NOTICE", 1).flatMap(lambda x : x.split(" ")).map(lambda x : (x, 1))\
.reduceByKey(lambda a, b : a + b).saveAsTextFile("/notebooks/wordCountNOTICE_homework3")
```

```python
# open file in read mode
    # read the file content into a string, split the string into a list of words separtated by "\n"
    # convert the list into set by removing duplicate words
with open("/notebooks/english_dictionary.txt", "r") as fp:
    wordList = fp.read().split("\n")
    wordSet = set(wordList)
```

```python
# print wordSet (unique words)
wordSet
```

```python
# return the number of elements in the set (number of unique words)
len(wordSet)
```

```python
# return the total number of elements in the list
len(wordList)
```

```python
# read file "/spark/NOTICE" and creates an RDD of the words, number of partitions is "1"
# flatMap makes the output a flat sequence of each word as an element
rddWord = sc.textFile("/spark/NOTICE", 1).flatMap(lambda x : x.split(" "))
```

```python
# total number of words in the RDD rddWord
rddWord.count()
```

```python
# creats an RDD, having words from "/spark/NOTICE" that exist in wordSet (including duplicates)
rddEnglishWord = rddWord.filter(lambda x : x in wordSet)
```

```python
# number of words in the RDD rddEnglishWord
rddEnglishWord.count()
```

```python
# create an RDD that contains tuples of (word, 1) for each word
rddInitWordPair = rddEnglishWord.map(lambda x : (x, 1))
```

```python
# create an RDD that contains tuples of (word, counts) for each unique word 
# counts is the sum of all counts for a unique word
rddWordCountPair = rddInitWordPair.reduceByKey(lambda a, b : a+b)
```

```python
# first 10 elements from rddWordCountPair
rddWordCountPair.take(10)
```
